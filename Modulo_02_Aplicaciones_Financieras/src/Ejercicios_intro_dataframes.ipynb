1. Crea un DataFrame llamado df que contenga los siguientes datos sobre productos:

```python
productos = ['Manzana', 'Banana', 'Naranja', 'Pera', 'Kiwi']
precios = [1.5, 0.5, 1.2, 1.0, 2.0]
cantidades = [100, 150, 200, 120, 80]
categorias = ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta']

# Ejercicio 1
import pandas as pd

df_productos = pd.DataFrame({
    'Producto': productos,
    'Precio': precios,
    'Cantidad': cantidades,
    'Categoría': categorias
})

df_productos
```

2. Acceso a Filas y Columnas
* Ejercicio 2.1: Accede a la primera fila del DataFrame df y muéstrala.
* Ejercicio 2.2: Accede a la columna Precio y muéstrala

```python
# Ejercicio 2
# Acceder a la primera fila
primera_fila = df_productos.iloc[0]

# Acceder a la columna Precio
columna_precio = df_productos["Precio"]

print(primera_fila,'\n', columna_precio)
```

3. Filtrar Datos: Filtra los productos cuyo precio sea mayor a 1.0 y muéstralos.

```python
# Ejercicio 3
productos_precio_mayor_1 = df_productos[df_productos["Precio"] > 1.0]
productos_precio_mayor_1
```

4. Calcular el Valor Total: Crea una nueva columna llamada Valor Total que calcule el valor total de cada producto (Precio × Cantidad)

```python
# Ejercicio 4
df_productos["Valor Total"] = df_productos["Precio"] * df_productos["Cantidad"]
df_productos
```

5. Estadísticas Básicas: Calcula y muestra la media del precio y la cantidad de los productos.

```python
# Ejercicio 5
media_precio = df_productos['Precio'].mean()
media_cantidad = df_productos["Cantidad"].mean()

print(f"Media precio: {media_precio}\nMedia cantidad: {media_cantidad}")
```

5.1. Añade dos productos de una nueva categoría

```python
# Crear un DataFrame adicional con dos nuevos productos de una categoría diferente
nuevos_productos = pd.DataFrame({
    'Producto': ['Lechuga', 'Tomate'],
    'Precio': [0.8, 1.1],
    'Cantidad': [50, 70],
    'Categoría': ['Verdura', 'Verdura']
})

# Añadir los nuevos productos al DataFrame original
df_productos = pd.concat([df_productos, nuevos_productos], ignore_index=True)
df_productos
```

6. Agrupar Datos: Agrupa los productos por Categoría y calcula la suma del Valor Total para cada categoría.

```python
# Ejercicio 6

# Asegurarnos de que la columna "Valor Total" esté actualizada
df_productos["Valor Total"] = df_productos["Precio"] * df_productos["Cantidad"]

# Agrupar por Categoría y calcular la suma del Valor Total para cada una
valor_total_por_categoria = df_productos.groupby("Categoría")["Valor Total"].sum()
valor_total_por_categoria
```

7. Ordenar Datos: Ordena el DataFrame por Precio de forma descendente y muéstralo.

```python
# Ejercicio 7
df_ordenado_precio = df_productos.sort_values(by="Precio", ascending=False)
df_ordenado_precio
```

8. Datos Faltantes: Crea un DataFrame df2 que contenga algunos valores faltantes (por ejemplo, usa np.nan para representar valores faltantes). Luego, usa el método .fillna() para rellenar esos valores con 0.

```python
# Ejercicio 8
import numpy as np

# Crear un DataFrame con algunos valores faltantes
df2 = pd.DataFrame({
    'Producto': ['Manzana', 'Banana', 'Naranja', 'Pera', 'Kiwi'],
    'Precio': [1.5, np.nan, 1.2, 1.0, np.nan],
    'Cantidad': [100, 150, np.nan, 120, 80]
})

# Rellenar los valores faltantes con 0
df2_filled = df2.fillna(0)
df2, df2_filled
```

10. Crear una Columna Condicional: Sobre df_productos añade una nueva columna llamada Descuento que sea igual a 10% del precio si la cantidad del producto es mayor a 150. Si no, el descuento será 0. Luego, calcula el Precio Final restando el descuento al precio original.

```python
# Ejercicio 10
# Añadir la columna "Descuento" en función de la cantidad
df_productos['Descuento'] = df_productos['Precio'].where(df_productos['Cantidad'] > 150, 0) * 0.1

# Calcular el "Precio Final" restando el descuento al precio original
df_productos['Precio Final'] = df_productos['Precio'] - df_productos['Descuento']

# Mostrar el DataFrame actualizado
df_productos
```

12. Ordenar y Seleccionar los 3 Productos más Caros
Ejercicio: Ordena el DataFrame por la columna Precio Final de forma descendente y selecciona los 3 productos más caros, mostrando su Producto, Precio Final, y Cantidad.

```python
# Ejercicio 12
# Ordenar el DataFrame por la columna "Precio Final" de forma descendente
df_top3_caros = df_productos.sort_values(by="Precio Final", ascending=False).head(3)

# Seleccionar las columnas deseadas
df_top3_caros_resultado = df_top3_caros[['Producto', 'Precio Final', 'Cantidad']]
df_top3_caros_resultado
```
