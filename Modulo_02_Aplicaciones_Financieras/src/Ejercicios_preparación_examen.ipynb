Descarga de Datos de Criptomonedas
* Descarga los precios de cierre ajustado de Bitcoin, Ethereum, Cardano, Solana y Polkadot desde el 1 de enero de 2021 hasta el 31 de diciembre de 2023 en un DataFrame df_crypto.

```python
import yfinance as yf

tickers1 = ['BTC-USD', 'ETH-USD', 'ADA-USD', 'SOL-USD', 'DOT-USD']
inicio = '2021-1-1'
final = '2023-12-31'

df_crypto = yf.download(tickers1, start=inicio, end=final)['Adj Close']

df_crypto.head()
```

Descarga de Datos de Startups Tecnológicas
* Descarga los precios de cierre ajustado de Rivian, Airbnb, Palantir, DoorDash y Snowflake desde el 1 de enero de 2021 hasta el 31 de diciembre de 2023 en un DataFrame df_tech.

```python
tickers2 = ['RIVN', 'ABNB', 'PLTR', 'DASH', 'SNOW']

df_tech = yf.download(tickers2, start = inicio, end = final)['Adj Close']

df_tech = df_tech.dropna()

df_tech.head()
```

Cartera con Pesos Diferentes
* Crea una cartera con los pesos [0.25, 0.15, 0.2, 0.2, 0.2] para Bitcoin, Ethereum, Rivian, Palantir y Snowflake. Calcula y compara la rentabilidad y volatilidad anualizadas de esta cartera frente a la cartera equilibrada.

Generación de Carteras Simuladas
* Usa las rentabilidades de Bitcoin y Ethereum para simular 1000 combinaciones de cartera con pesos que varíen entre 0% a 100% para cada uno, y grafica la frontera eficiente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Filtramos las columnas de Bitcoin y Ethereum.
rent_log_data_btc_eth = rent_log_data_assets[['BTC-USD', 'ETH-USD']]

# Definir el número de simulaciones
n_simulaciones = 1000

# Arrays para almacenar rentabilidad y volatilidad de cada combinación de pesos
rentabilidades = []
volatilidades = []

# Simulación de 1000 combinaciones de pesos
for w_btc in np.linspace(0, 1, n_simulaciones):
    w_eth = 1 - w_btc
    pesos = np.array([w_btc, w_eth])

    # Calcular rentabilidad y volatilidad anualizadas de la cartera
    rentabilidad_cartera = (rent_log_data_btc_eth * pesos).sum(axis=1).mean() * 252
    volatilidad_cartera = (rent_log_data_btc_eth * pesos).sum(axis=1).std() * np.sqrt(252)

    # Almacenar los resultados
    rentabilidades.append(rentabilidad_cartera)
    volatilidades.append(volatilidad_cartera)

# Graficar la frontera eficiente
plt.figure(figsize=(10, 6))
plt.scatter(volatilidades, rentabilidades, c=rentabilidades, cmap='viridis', marker='o')
plt.colorbar(label='Rentabilidad Anualizada')
plt.title("Frontera Eficiente entre Bitcoin y Ethereum")
plt.xlabel("Volatilidad Anualizada")
plt.ylabel("Rentabilidad Anualizada")
plt.grid(True)
plt.show()
```

Frontera Eficiente para Startups
* Usa las rentabilidades de Rivian, DoorDash y Snowflake para generar y graficar 1000 combinaciones de cartera. Calcula y muestra la frontera eficiente entre estas startups.

```python
import numpy as np
import matplotlib.pyplot as plt

# Filtramos solo las columnas de Rivian, DoorDash y Snowflake.
rent_log_data_riv_dash_snow = rent_log_data_assets[['RIVN', 'DASH', 'SNOW']]

# Definir el número de simulaciones
n_simulaciones = 1000

# Arrays para almacenar rentabilidad y volatilidad de cada combinación de pesos
rentabilidades = []
volatilidades = []

# Simulación de combinaciones de cartera
for _ in range(n_simulaciones):
    # Generar pesos aleatorios y normalizar para que sumen 1
    pesos = np.random.random(3)
    pesos /= pesos.sum()

    # Calcular rentabilidad y volatilidad anualizadas de la cartera
    rentabilidad_cartera = (rent_log_data_riv_dash_snow * pesos).sum(axis=1).mean() * 252
    volatilidad_cartera = (rent_log_data_riv_dash_snow * pesos).sum(axis=1).std() * np.sqrt(252)

    # Almacenar los resultados
    rentabilidades.append(rentabilidad_cartera)
    volatilidades.append(volatilidad_cartera)

# Graficar la frontera eficiente
plt.figure(figsize=(10, 6))
plt.scatter(volatilidades, rentabilidades, c=rentabilidades, cmap='viridis', marker='o')
plt.colorbar(label='Rentabilidad Anualizada')
plt.title("Frontera Eficiente entre Rivian, DoorDash y Snowflake")
plt.xlabel("Volatilidad Anualizada")
plt.ylabel("Rentabilidad Anualizada")
plt.grid(True)
plt.show()
```

Evaluación de Riesgo y Retorno de Carteras
* Compara las rentabilidades y volatilidades anualizadas de las carteras formadas por Bitcoin-Ethereum y Palantir-Rivian en un gráfico, utilizando diferentes combinaciones de pesos y visualizando la relación riesgo-retorno de ambas combinaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Filtrar las columnas correspondientes a Bitcoin y Ethereum, y Palantir y Rivian
rent_log_btc_eth = rent_log_data_assets[['BTC-USD', 'ETH-USD']]
rent_log_pal_riv = rent_log_data_assets[['PLTR', 'RIVN']]

# Definir el número de combinaciones de pesos
n_combinaciones = 100

# Arrays para almacenar rentabilidad y volatilidad para cada cartera
rentabilidades_btc_eth = []
volatilidades_btc_eth = []

rentabilidades_pal_riv = []
volatilidades_pal_riv = []

# Generar combinaciones de pesos para ambas carteras
for w_btc in np.linspace(0, 1, n_combinaciones):
    w_eth = 1 - w_btc
    pesos_btc_eth = np.array([w_btc, w_eth])

    # Calcular rentabilidad y volatilidad anualizadas de la cartera Bitcoin-Ethereum
    rentabilidad_btc_eth = (rent_log_btc_eth * pesos_btc_eth).sum(axis=1).mean() * 252
    volatilidad_btc_eth = (rent_log_btc_eth * pesos_btc_eth).sum(axis=1).std() * np.sqrt(252)
    rentabilidades_btc_eth.append(rentabilidad_btc_eth)
    volatilidades_btc_eth.append(volatilidad_btc_eth)

    # Generar pesos para la cartera Palantir-Rivian
    w_pal = w_btc
    w_riv = 1 - w_btc
    pesos_pal_riv = np.array([w_pal, w_riv])

    # Calcular rentabilidad y volatilidad anualizadas de la cartera Palantir-Rivian
    rentabilidad_pal_riv = (rent_log_pal_riv * pesos_pal_riv).sum(axis=1).mean() * 252
    volatilidad_pal_riv = (rent_log_pal_riv * pesos_pal_riv).sum(axis=1).std() * np.sqrt(252)
    rentabilidades_pal_riv.append(rentabilidad_pal_riv)
    volatilidades_pal_riv.append(volatilidad_pal_riv)

# Graficar la relación riesgo-retorno de ambas combinaciones de carteras
plt.figure(figsize=(10, 6))
plt.scatter(volatilidades_btc_eth, rentabilidades_btc_eth, label="Cartera Bitcoin-Ethereum", c='blue', alpha=0.6)
plt.scatter(volatilidades_pal_riv, rentabilidades_pal_riv, label="Cartera Palantir-Rivian", c='green', alpha=0.6)
plt.xlabel("Volatilidad Anualizada")
plt.ylabel("Rentabilidad Anualizada")
plt.title("Comparación de Riesgo-Retorno de las Carteras Bitcoin-Ethereum y Palantir-Rivian")
plt.legend()
plt.grid(True)
plt.show()
```
