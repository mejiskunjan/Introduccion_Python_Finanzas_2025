## Ejercicio 8 (1 punto)

**Utilizando `data_rentab_log`, crea una cartera equiponderada de todos los activos y calcula su rentabilidad y volatilidad anualizada.**

```python
# Creo los pesos de la cartera equiponderada.
w = np.ones(data_rentab_log.shape[1])
w = np.array(w / np.sum(w))  # (6,)

# Calculo la rentabilidad de la cartera.
r_p = data_rentab_log @ w  # (271, 6) @ (6,) = (271,)

# Calculo la volatilidad de la cartera.
# Opción 1
# sigma_p = np.std(r_p)
# Opción 2
Sigma = data_rentab_log.cov()  # (6, 6)
sigma_p = np.sqrt(w.T @ Sigma @ w)  # (,6) @ (6, 6) @ (6,) = (1,)

# Calculo la rentabilidad y volatilidad anualizadas.
r_p_anualizada = r_p.mean() * 252
sigma_p_anualizada = sigma_p * np.sqrt(252)

# Imprimo resultados.
print(f"Rentabilidad anualizada = {r_p_anualizada:.2%}")
print(f"Volatilidad anualizada = {sigma_p_anualizada:.2%}")
```

## Ejercicio 9 (1 punto)

**Utilizando `data_rentab_log`, grafica el histograma de las rentabilidades de Google y compáralo con una distribución normal con la misma media y desviación típica que la distribución de Google.**

```python
from numpy import random

# Seleccionamos los datos de rentabilidades de Google
data_google = data_rentab_log.loc[:, ["GOOGL"]]
# Calculamos su media.
mu = data_google.mean()
# Calculamos su desviación típica.
sigma = data_google.std()

# Creamos el generador de números aleatorios.
rng = random.default_rng(42)
# Creamos el mismo número de datos siguiendo una distribución normal con la media y desviación típica de Google.
data_normal = rng.normal(loc=mu, scale=sigma, size=len(data_google))

# Graficamos las dos series.
plt.figure(figsize=(10, 5))
plt.hist(data_google, bins=50, label="Google")
plt.hist(data_normal, bins=50, alpha=0.5, label="Distribución Normal")
plt.title("Rentabilidades de GOOGL vs Normal")
plt.xlabel("Rentabilidad")
plt.ylabel("Repeticiones")
plt.legend()
plt.grid(True)
plt.show()
```

## Ejercicio 10 (1 punto)

**Utilizando `data_rentab_log` y solamente las rentabilidades de Tesla y Telefónica, crea 1000 carteras cuyos pesos vayan desde 0% para Tesla y 100% para Telefónica, hasta 100% para Tesla y 0% para Telefónica. Calcula las rentabilidades y volatilidades anualizadas para todas las carteras y grafícalas situando la volatilidad en el eje X y la rentabilidad en el eje Y.**

```python
# Filtro las rentabilidades de los activos que me interesan.
rentab_activos = data_rentab_log.loc[:, ["TSLA", "TEF.MC"]]

# Para cada cartera... partiendo del peso del primer activo...
rentabilidades_anualizadas = []
volatilidades_anualizadas = []
for w1 in np.linspace(start=0, stop=1, num=1000):
    # Calculo el peso del segundo activo.
    w2 = 1 - w1
    # Creo el vector de pesos.
    w = np.array([w1, w2])

    # Calculo la rentabilidad de la cartera.
    r_p = rentab_activos @ w  # (271, 2) @ (2,) = (271,)

    # Calculo la volatilidad de la cartera.
    # Opción 1
    # sigma_p = np.std(r_p)
    # Opción 2
    Sigma = rentab_activos.cov()  # (2, 2)
    sigma_p = np.sqrt(w.T @ Sigma @ w)  # (,2) @ (2, 2) @ (2,) = (1,)

    # Calculo la rentabilidad y volatilidad anualizadas.
    r_p_anualizada = r_p.mean() * 252
    sigma_p_anualizada = sigma_p * np.sqrt(252)

    # Guardo los resultados.
    rentabilidades_anualizadas.append(r_p_anualizada)
    volatilidades_anualizadas.append(sigma_p_anualizada)

# Grafico las rentabilidades y volatilidades de las carteras.
plt.figure(figsize=(10, 5))
plt.scatter(volatilidades_anualizadas, rentabilidades_anualizadas)
plt.title("Frontera eficiente")
plt.xlabel("Volatilidad")
plt.ylabel("Rentabilidad")
plt.grid(True)
plt.show()
```
